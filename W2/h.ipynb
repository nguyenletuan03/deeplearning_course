```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(10,8)
plt.rcParams['figure.dpi']=150
plt.rcParams['font.size']=16
```

```python
def cost_func(theta):
    return theta**2-4*theta+8

def daoham(theta):
    return 2*theta-4

def gra_desent(start_theta, learning_rate, daoham, nguongdung):
    theta= start_theta
    theta_his=[theta]
    cost_his=[cost_func(theta)]
    while True:
        theta=theta-(learning_rate * daoham(theta))
        theta_his.append(theta)
        cost_his.append(cost_func(theta))
        if abs(daoham(theta))<nguongdung:
            break
    return theta_his, cost_his
```

```python
start_theta= 0
learning_rate=0.1
nguongdung=0.0001

theta_his, cost_his= gra_desent(start_theta, learning_rate, daoham, nguongdung)

theta=np.linspace(-1,5,400)
cost_values= cost_func(theta)
```

```python
plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.plot(theta, cost_values, label='$J(\\theta)=\\theta^2-4\\theta +8$', color='darkorange')
plt.scatter(theta_his,cost_his,color='darkorange', linestyle='--',lw=2)

plt.xlabel('parameters-$\\theta$')
plt.ylabel('Cost-$J(\\theta)$')
plt.title('Cost function and gradient desent path')
plt.legend()

plt.subplot(1,2,2)
plt.plot(cost_his, label='Cost History',color='darkorange', lw=2)
plt.xlabel('Iteration')
plt.ylabel('Cost-$J(\\theta)$')
plt.title('Cost history during gradient desent')
plt.legend()

plt.tight_layout()
plt.show()
```
