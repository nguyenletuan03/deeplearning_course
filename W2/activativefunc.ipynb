```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize']=(10,8)
plt.rcParams['figure.dpi']=150
plt.rcParams['font.size']=16
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

x=np.linspace(-10,10,400)
y=sigmoid(x)
```

```python
plt.plot(x,y, label='Sigmoid Function',color='darkorange',lw=8)
plt.title('Sigmoid Activation Function', fontweight='bold')
plt.xlabel('input',fontweight='bold')
plt.ylabel('output',fontweight='bold')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def tanh(x):
    return np.tanh(x)

x=np.linspace(-10,10,400)
y=tanh(x)
```

```python
plt.plot(x,y, label='Sigmoid Function',color='darkorange',lw=8)
plt.title('Tanh Activation Function', fontweight='bold')
plt.xlabel('input',fontweight='bold')
plt.ylabel('output',fontweight='bold')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def relu(x):
    return np.maximum(0,x)

x=np.linspace(-10,10,400)
y=relu(x)
```

```python
plt.plot(x,y, label='Sigmoid Function',color='darkorange',lw=8)
plt.title('ReLU Activation Function', fontweight='bold')
plt.xlabel('input',fontweight='bold')
plt.ylabel('output',fontweight='bold')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def leaky_relu(x, alpha=0.01):
    return np.where(x>0,x,x*alpha)

x=np.linspace(-10,10,400)
y=leaky_relu(x)
```

```python
plt.plot(x,y, label='Sigmoid Function',color='darkorange',lw=8)
plt.title('Leaky ReLU Activation Function', fontweight='bold')
plt.xlabel('input',fontweight='bold')
plt.ylabel('output',fontweight='bold')
plt.grid(True)
plt.legend()
plt.show()
```

```python
def softmax(x):
    e_x = np.exp(x - np.max(x))
    A = e_x / e_x.sum()
    return A

x=np.linspace(-10,10,400)
y=softmax(x)
```

```python
plt.plot(x,y, label='Sigmoid Function',color='darkorange',lw=8)
plt.title('Softmax Activation Function', fontweight='bold')
plt.xlabel('input',fontweight='bold')
plt.ylabel('output',fontweight='bold')
plt.grid(True)
plt.legend()
plt.show()
```
