```python
import tensorflow as tf
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
matrix=np.array([[50,50,50,0,0,0],
[50,50,50,0,0,0],
[50,50,50,0,0,0],
[0,0,0,50,50,50],
[0,0,0,50,50,50],
[0,0,0,50,50,50]], dtype=np.float32)
```

```python
filter_= np.array([[1,0,-1],
[1,0,-1],
[1,0,-1]], dtype=np.float32)
```

```python
matrix=matrix.reshape((1,6,6,1))#shape: batch_size, height, width, channels
filter_=filter_.reshape((3,3,1,1)) #shape:(filter_height, filter width, in channels, out channel)
```

```python
matrix
```

```python
conv=tf.nn.conv2d(matrix, filter_, strides=[1,1,1,1], padding='VALID')
result =conv.numpy()
result= result.reshape((result.shape[1], result.shape[2]))
```

```python
fix, axes=plt.subplots(1,3, figsize=(18,6))
axes[0].imshow(matrix.reshape((6,6)), cmap='gray', interpolation='nearest')
axes[0].set_title('Grayscale image (input matrix)')
axes[0].axis('off')

axes[1].imshow(filter_.reshape((3,3)), cmap='gray', interpolation='nearest')
axes[1].set_title('Filter(Kernel)')
axes[1].axis('off')

axes[2].imshow(result, cmap='gray', interpolation='nearest')
axes[2].set_title('Convolution Result')
axes[2].axis('off')
```
